
3. X l tn ct tn Name, tch ra lm 2 ct: firstName v secondName.

```python
#Tch ct name
df[['firstName', 'secondName']]=df['Name'].str.split(',',expand = True)
df.head(5)
```

```python
#Xo ct Name
df = df.drop('Name',axis=1)
df.head(5)
```

4. X l rt gn kch thc d liu trn ct Sex nh sau: thay th male  M v female  F



```python
Sex = df['Sex']
for i in range(len(Sex)):
  if  Sex[i] == 'male':
    df.Sex[i] = 'M'
  if  Sex[i] == 'female':
    df.Sex[i] = 'F'
df.head(5)
```

5. X l d liu thiu trn bin Age bng cch thay th bng gi tr trung bnh tui: Hy a ra quyt nh dng gi tr trung bnh tui ton b hnh khch hay theo tng nhm hng v (hng hnh khch: Pclass).

```python
sns.boxplot(y='Age', x='Pclass', data = df)
```

```python
# Nhn xt v tui trung bnh gia cc nhm hnh khch.
#==> Phn phi tui trn hng Pclass = 1 c dng chun hn so vi hng 2 v hng 3

#Tin hnh thay th gi tr Age b thiu.
# d liu thiu trn bin Age bng cch thay th bng gi tr trung bnh tui
df['Age'].fillna(df['Age'].mean(), inplace =True)
#Hin th kt qu dng bng 
df
```

```python
#Trc quan d liu  x l thiu cho ctAge bng biu  Heat map.
sns.heatmap(df[['Age']].isna(), cmap = 'Reds', yticklabels= False)
```

6. Xy dng bin s Agegroup c thang o th t c nh x theo thang o khong da trn  tui ca hnh khch nh sau: (age =< 12]  Kid; (12, 18]: Teen, (18, 60]: Adult v ( age > 60): Older

```python
for i in df['Age']:
  df.loc[(df['Age'] <=  12), 'Agegroup'] ='Kid'
  df.loc[(df['Age'] > 12) & (df['Age'] <= 18), 'Agegroup'] ='Teen'
  df.loc[(df['Age'] > 18) & (df['Age'] <= 60), 'Agegroup'] ='Adult'
  df.loc[(df['Age'] > 60), 'Agegroup'] ='Older'
df.head(10)
```


7. Tin hnh thm c trng v danh xng (namePrefix) trong x hi bng cch tch Mr, Mrs, Miss, Master ra khi secondName

```python
# tch ct secondName
df[['namePrefix','secondName' ]] =df['secondName'].str.split('. ',1,expand = True)
df.head(5)
```

8. Khai thc thm thng tin s lng thnh vin i theo nhm thn quen (familySize) i vi mi hnh khch trn chuyn hi trnh; family size = 1+ SibSp + Parch

```python
df['familySize'] = 1+df.SibSp+df.Parch
df.head(5)
```


9. To thm c trng Alone  xc nh hnh khch i theo nhm hay c nhn bng cch da trn familySize nh sau: Nu familySize = 0 th gi tr Alone = 1 v ngc li l 0.

```python
for i in df['familySize']:
  df.loc[(df['familySize'] ==  0), 'Alone'] = 1
  df.loc[(df['familySize'] !=  0), 'Alone'] = 0
df.Alone
```

10. Tin hnh tch loi cabin (typeCabin) m hnh khch   lc v phn tch c tnh cabin. Loi cabin c k hiu bi ch ci u tin. Lu : i vi d liu cabin b thiu th thay th bng Unknown

```python
df['typeCabin']= None
Cabin= df.Cabin
for i in range(len(Cabin)):
  if pd.isna(df.Cabin[i]) == True:
    df['typeCabin'][i]= 'Unknown'
  else:
    df['typeCabin'][i]= df.Cabin[i][0]
```
