```python

# Khai báo các thư viện
import numpy
import pandas as pd
from scipy import stats
import seaborn as sns
```

Mean - Mode - Median
Mode: là giá trị xuất hiện nhiều nhất trong tập dữ liệu

Ví dụ 1:

```python
speed = [1,6,7,8,11,6,13,8,4,78,7,8,8]
mode = stats.mode(speed)

print(mode)
```

Median Giá trị ở giữa khi bạn đã sắp xếp hết các giá trị

Ví dụ 2

```python
speed =[1,6,7,8,11,6,13,8,4,78,7,8,8]
median = numpy.median(speed)

print(median)
```

Mean

Giá trị trung bình. Để tính giá trị trung bình, hãy tìm tổng của tất cả các giá trị và chia tổng cho số giá trị

Ví dụ 3

```python
speed = [1,6,7,8,11,6,13,8,4,78,7,8,8]

mean = numpy.mean(speed)

print(mean)
```

Thí dụ 4: Dùng đồ thị Histogram, đồ thị boxplot và các độ đo thống kê để khảo sát hình dạng phân phối

```python
# lấy dữ liệu
df = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv")
```

```python
sns.displot(df, x="T1", kde=True)
```

```python

sns.boxplot(x='T1', data = df)
```


Dựa vào histogram ta có thể thấy hình dạng phân phối của T1 có dạng xấp xỉ đối xứng, nhưng đồ thị boxplot ta lại thấy phân phối của T1 hơi lệch trái. Ta dùng thêm các đặc trưng số của T1 để có thêm thông tin

```python
mean = df['T1'].mean()
mode = df['T1'].mode()
median = df['T1'].median()
print ('mean= ', mean)
print ('median= ', median)
```

```python

print ('mode= ', mode)
```


Giá trị mode chỉ là 5.2; 5.3; 7.3. Biến T1 là biến ngẫu nhiên nhận các giá trị liên tục, nên Mode ở đây được hiểu là khoảng chứa dữ liệu nhiều nhất, cụ thể mode ở đây phải thuộc khoảng từ 5.2 đến 7.3. Cho nên ta vẫn có thể dựa vào Mean, Media để đưa ra kết luận là hình dáng phân phối của biến T1.

Ta thấy Mean > Median nên hình dáng phân phối của T1 là lệch phải

Skewness và Kurtosis
Skewness: "Skewness đánh giá mức độ mà phân phối của một biến là đối xứng. Nếu phân phối các phản hồi cho một biến trải dài về phía đuôi bên phải hoặc bên trái của phân phối, thì phân phối được gọi là lệch

Kurtosis là thước đo liệu phân bố có quá đỉnh hay không (phân bố rất hẹp với hầu hết các phản hồi ở trung tâm).

```python
#Ví dụ:
#calculate sample skewness
from scipy.stats import skew
skew(df['T1'], bias=False)
```


Đối với kurtosis, nguyên tắc chung là nếu số lượng lớn hơn +1, phân phối quá đỉnh. Tương tự như vậy, kurtosis nhỏ hơn –1 cho thấy phân phối quá bằng phẳng. Các phân phối có biểu hiện lệch và / hoặc kurtosis vượt quá các hướng dẫn này được coi là không bình thường.
