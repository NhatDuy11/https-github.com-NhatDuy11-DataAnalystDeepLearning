Th d 4: Dng  th Histogram,  th boxplot v cc  o thng k  kho st hnh dng phn phi

```python
# ly d liu
df = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/dulieuxettuyendaihoc.csv")
```

```python
sns.displot(df, x="T1", kde=True)
```

```python

sns.boxplot(x='T1', data = df)
```


Da vo histogram ta c th thy hnh dng phn phi ca T1 c dng xp x i xng, nhng  th boxplot ta li thy phn phi ca T1 hi lch tri. Ta dng thm cc c trng s ca T1  c thm thng tin

```python
mean = df['T1'].mean()
mode = df['T1'].mode()
median = df['T1'].median()
print ('mean= ', mean)
print ('median= ', median)
```

```python

print ('mode= ', mode)
```
